```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
X, y = load_wine(return_X_y = True)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101, shuffle = True, stratify = y)
```

```python
print('The shape of training dataset is:', X_train.shape)
print('The shape of test dataset is:', X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
lin_svc = SVC(kernel = 'linear')
```

```python
rbf_svc = SVC(kernel = 'rbf')
```

```python
poly_svc = SVC(kernel = 'poly', degree = 2)
```

```python
lin_svc.fit(X_train_scaled, y_train)
```

```python
rbf_svc.fit(X_train_scaled, y_train)
```

```python
poly_svc.fit(X_train_scaled, y_train)
```

```python
lin_sc = lin_svc.score(X_test_scaled, y_test)
```

```python
print('The accuracy of svc model with linear kernel is:', lin_sc)
```

```python
rbf_sc = rbf_svc.score(X_test_scaled, y_test)
```

```python
print('The accuracy of svc model with rbf kernel is:', rbf_sc)
```

```python
poly_sc = poly_svc.score(X_test_scaled, y_test)
```

```python
print('The accuracy of svc model with polynomial kernel is:', poly_sc)
```

```python
lin_pred = lin_svc.predict(X_test_scaled)
```

```python
rbf_pred = rbf_svc.predict(X_test_scaled)
```

```python
poly_pred = poly_svc.predict(X_test_scaled)
```

```python
print('Classification report for svc model with linear kernel is:\n')
print(classification_report(y_test, lin_pred))
```

```python
print('Classification report for svc model with rbf kernel is:\n')
print(classification_report(y_test, rbf_pred))
```

```python
print('Classification report for svc model with polynomial kernel is:\n')
print(classification_report(y_test, poly_pred))
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(confusion_matrix(y_test, lin_pred), annot = True, square = True, cbar = True, annot_kws = {'size': 15}, fmt = 'd', 
            cmap = 'Blues')
plt.title('Confusion matrix for svc model with linear kernel', fontsize = 15)
plt.xlabel('Predicted', fontsize = 10)
plt.ylabel('Truth', fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(confusion_matrix(y_test, rbf_pred), annot = True, square = True, cbar = True, annot_kws = {'size': 15}, fmt = 'd',
            cmap = 'Blues')
plt.title('Confusion matrix for svc model with rbf kernel', fontsize = 15)
plt.xlabel('Predicted', fontsize = 10)
plt.ylabel('Truth', fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(confusion_matrix(y_test, poly_pred), annot = True, square = True, cbar = True, annot_kws = {'size': 15}, fmt = 'd', 
            cmap = 'Blues')
plt.title('Confusion matrix for svc model with polynomial kernel', fontsize = 15)
plt.xlabel('Predicted', fontsize = 10)
plt.ylabel('Truth', fontsize = 10)
plt.show()
```
